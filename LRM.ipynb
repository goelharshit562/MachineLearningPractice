```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('50_Startups.csv')
df.head()
```

```python
dumy=pd.get_dummies(df.State)
dumy
xy=dumy.drop(['New York'],axis=1)
merge=pd.concat([df,xy],axis=1)
final=merge.drop(['State'],axis=1)
x=final.drop(['Profit'],axis=1)
x

```

```python
y=final.Profit.values
y=y.reshape(-1,1)
y


```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=0)

sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

y_train=sc.fit_transform(y_train)
y_test=sc.transform(y_test)
x_test=sc.transform(x_test)
lm.fit(x_train,y_train)
pred=lm.predict(x_test)
```

```python
df1=pd.DataFrame({'Actual':y_test,'predicted':pred})
df2=df1.head(5)
df2
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_test,pred))
```

```python
x=df.iloc[:,:1].values
y=df.iloc[:,-1:].values
x
```

```python

def gd(x,y):
    m_cur=c_cur=0
    itera=1000
    n=len(x)
    learning_rate=0.01
    
    for i in range(itera):
        y_predict=m_cur*x+c_cur
        cost=(1/n)*sum(val**2 for val in (y-y_predict))
        md= -(2/n)*sum(x*(y-y_predict))
        cd= -(2/n)*sum(y-y_predict)
        m_cur=m_cur - learning_rate * md
        c_cur=c_cur - learning_rate * cd
        print("m {} , c {}, cost {},pred{}, iteration {}".format(m_cur,c_cur,cost,y_predict,i))

gd(x_train,y_train)
                       
                   
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(x_test,y)
```
